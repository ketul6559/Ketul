```python
X = df.drop(labels = 'status', axis = 1)
y = df['status']
```

```python
# Split the dataset into training and test set in the ratio of 70:30
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 859, stratify = y)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion = 'entropy')
dt.fit(X_train,y_train)
y_predict = dt.predict(X_test)
```

```python
from sklearn import metrics
print("Training score:-", dt.score(X_train, y_train))
print("Testing Score:-", dt.score(X_test, y_test))
print(metrics.confusion_matrix(y_test, y_predict))
```

### With parameters

```python
dt = DecisionTreeClassifier(criterion = 'entropy', random_state = 859, max_depth = 50, min_samples_leaf = 7)
dt.fit(X_train,y_train)
y_predict = dt.predict(X_test)
```

```python
print("Training score:-", dt.score(X_train, y_train))
print("Testing Score:-", dt.score(X_test, y_test))
print(metrics.confusion_matrix(y_test, y_predict))
```

```python
print(metrics.classification_report(y_test, y_predict))
```

```python
metrics.accuracy_score(y_test, y_predict)
```

Without parameters, we can see the decision tree is overfitted model and accuracy score is low.

With regularization, the model becomes excellent and fits perfectly, even the testing score is excellent so regularization indeed improves the algorithm.

In the classification report, precision, recall, and F1 score are quite good.

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rm = RandomForestClassifier(n_estimators = 50)
rm.fit(X_train, y_train)
y_predict = rm.predict(X_test)
```

```python
print("Training score:-", rm.score(X_train, y_train))
print("Testing Score:-", rm.score(X_test, y_test))
print(metrics.confusion_matrix(y_test, y_predict))
```

### With parameters 

```python
rm = RandomForestClassifier(n_estimators = 50, random_state = 4486, max_depth = 50, min_samples_leaf = 5)
rm.fit(X_train, y_train)
y_predict = rm.predict(X_test)
```

```python
print("Training score:-", rm.score(X_train, y_train))
print("Testing Score:-", rm.score(X_test, y_test))
print(metrics.confusion_matrix(y_test, y_predict))
```

```python
print(metrics.classification_report(y_test, y_predict))
```

```python
metrics.accuracy_score(y_test, y_predict)
```

Without parameters, the result is the same as the Decision Tree, overfitted model overall.

When adding parameters, the result is very excellent. The model fits perfectly and the testing accuracy is also increased. Hence adding parameters improve the algorithm. Even precision, recall, and F1 score is better than the decision tree.

## Conclusion 

While using two algorithms, both have good results, but when come to accuracy, the random forest algorithm performs very well over the decision tree. As we saw a correlation with each attribute, most of were strong positive linear relationships thus, with the help of this, we got more accurate results, and outliers didn't affect much to the dataset. So for this dataset, Random Forest did an excellent job.
