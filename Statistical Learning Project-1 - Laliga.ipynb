```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Laliga.csv')
df
```

```python
df = df.rename(columns = df.iloc[0]) # remove unnnamed columns and replaced with original name
df = df.drop(df.index[0])
df
```

```python
df = df.replace('-', 0) # replace dashes with 0 
df
```

```python
df['Debut'] = df['Debut'].astype(str).str[:4].astype(int)
```

```python
df['Debut']
```

```python
df[df['Debut'].between(1930, 1980)] # Team which have played between year from 1930 to 1980
```

```python
df['Points'] = df['Points'].astype(str).astype(int)
```

```python
df.info()
```

```python
df.sort_values(by = 'Points', ascending = False).head(5) # Top 5 team terms in point
```

```python
df['GoalsFor'] = df['GoalsFor'].astype(str).astype(int)
df['GoalsAgainst'] = df['GoalsAgainst'].astype(str).astype(int)
```

```python
df.info()
```

```python
def Goal_diff_count():
    return df['GoalsFor'] - df['GoalsAgainst']
```

```python
Goal_diff_count() # maximum and minimum goal difference
print('Max:-',Goal_diff_count().max())
print('Min:-',Goal_diff_count().min())
```

```python
Winning_Percent = Goal_diff_count() # name “Winning Percent” and append it to the data set
df['Winning Percent'] = Winning_Percent
df
```

```python
df['GamesWon'] = df['GamesWon'].astype(str).astype(int)
df['GamesPlayed'] = df['GamesPlayed'].astype(str).astype(int)
df.info()
```

```python
Percentage_of_Winning = (df['GamesWon'] / df['GamesPlayed']) * 100
Percentage_of_Winning
```

```python
df['Percentage_of_Winning'] = Percentage_of_Winning
df = df.sort_values('Percentage_of_Winning', ascending = False)
df['Team'].head(5) # top 5 teams which have the highest Winning percentage
```

```python
df['BestPosition'] = df['BestPosition'].astype(str).astype(int)
df.info()
```

```python
best_position = df.groupby(by = 'BestPosition') # Group teams based on their “Best position”
best_position.sum()
```
